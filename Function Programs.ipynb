Write a python code which checks whether the input number is a Perfect Square of not using a function.

```python
import math as m
def prefect_sqrt(n):
    sq = int(m.sqrt(n))
    if sq*sq == n:
        return 1
n = int(input("Enter the number: "))
check = prefect_sqrt(n)
if check == 1:
    print(n," is a perfect square")
else:
    print(n," is not a perfect square")
```

If the customer is a senior citizen (age > 60), fix the rate of interest (R) at 12%, if the customer is “female,” fix the rate of interest (R) at 10%; and for others, fix the rate of interest (R) at 9%.

**The function should take P, N, and R values as parameters.**


```python
def interest_calculation(p,n,r):
    return (p*n*r)/100
    
name = input("Enter the name: ")
age = int(input("Enter your age: "))
gen = input("Your Gender (Male = M and Female = F): ")
p = int(input("Enter your principle amount: "))
n = int(input("Enter the number of years: "))

if age > 60:
    print("Simple Interest will be ",interest_calculation(p,n,12)," at rate of interest as 12%")
elif gen == 'f' or gen == 'F':
    print("Simple Interest will be ",interest_calculation(p,n,10)," at rate of interest as 10%")
else:
    print("Simple Interest will be ",interest_calculation(p,n,9)," at rate of interest as 9%")
```

The below picture includes five geometrical shapes. 

You have to write Python code that includes five different functions: sphere(), cylinder(), cone(), rectangular_prism(), and triangular_prism(). You should get the respective values in the formula for the shapes, and they should be passed to the function as a parameter. If any of the shapes include **"π” (pi)**, then pi value should be passed as the default argument, and functions should **return the SURFACE AREA AND VOLUME** of the respective geometrical shapes. 

The formula for calculating the surface area and volume of the shapes is given in the picture. 


```python
def sphere(r,pi=3.14):
    sa = 4*pi*r**2
    v = (4/3)*pi*r**3
    return sa,v

def cylinder(r,h,pi=3.14):
    sa = 2*pi*r**2 + 2*pi*r*h
    v = pi*r**2*h
    return sa,v

def cone(r,h,s,pi=3.14):
    sa = pi*r*s + pi*r**2
    v = (1/3)*pi*r**2*h
    return sa,v

def rect_prism(l,h,w):
    sa = 2*(l*w+l*h+w*h)
    v = l*w*h
    return sa,v

def tri_prism(l,b,h,s):
    sa = b*h+2*l*s+l*b
    v = (1/2) * b * l * h
    return sa,v

print("1.Sphere\n2.Cylinder\n3.Cone\n4.Rectangular Prism\n5.Triangular Prism")
ch = int(input("Enter your choice : "))
if ch == 1:
    r = float(input("Enter the radius: "))
    s,v = sphere(r)
    print("Surface Area =",s,"and Volume of Sphere =",v)
elif ch == 2:
    r = float(input("Enter the radius: "))
    h = float(input("Enter the height: "))
    s,v = cylinder(r,h)
    print("Surface Area =",s,"and Volume of Cylinder =",v)
elif ch == 3:
    r = float(input("Enter the radius: "))
    h = float(input("Enter the height: "))
    s = float(input("Enter the side: "))
    s,v = cone(r,h,s)
    print("Surface Area =",s,"and Volume of cone =",v)
elif ch == 4:
    l = float(input("Enter the length: "))
    h = float(input("Enter the height: "))
    w = float(input("Enter the width: "))
    s,v = rect_prism(l,h,w)
    print("Surface Area =",s,"and Volume of Rectangular Prism =",v)
elif ch == 5:
    l = float(input("Enter the length: "))
    h = float(input("Enter the height: "))
    b = float(input("Enter the base: "))
    s = float(input("Enter the side: "))
    s,v = tri_prism(l,b,h,s)
    print("Surface Area =",s,"and Volume of Triangular Prism =",v)
else:
    print("Invalid Input")
    
```

Write a Python code to find the nth term of the Fibonacci series using RECURSION.

Sample Input:
5

Sample Output:
3


```python
#fibonacci series generation using for loop 
def fibo_for(n):
    a = 0
    b = 1
    print(a,b,end = ",",sep=",")
    for i in range(2,n):
        c = a+b
        print(c,end=",")
        a = b
        b = c
#To display the nth term of the Fibonacci series 
def fibo(n):
    if n == 0 or n == 1:
        return n
    else:
        return fibo(n-1)+fibo(n-2)
#fibonacci series generation using for recursion
def fibo_rec(n):
    if n == 0:
        return n
    elif n == 1:
        return n
    else:
        return fibo(n-1)+fibo(n-2)
    
n = int(input("Enter the n value to find the term: "))

print(n,"th term of the fibonacci series is ",fibo(n)) #--> will print the nth term in the series

#if we need to print the series within a given range, then we have to call the function inside the loop
e = int(input("Enter the n value of the series, for displaying the series from 0 to n: "))
for i in range(e+1):  #--> will start displaying from the 1st term(0) till the n values
    print(fibo_rec(i),end=",")
print("\b")
```

Write a Python function that takes base and power as arguments and calculates the power of numbers. 

Sample Input:

To Calculate 5 to the power of 3 then, enter the input as

Base : 5<br>
Power : 3

Sample Output:<br>
125

```python
def pow_rec(b,p):
    if p == 1:
        return b
    else:
        return b * pow_rec(b,p-1)

base = int(input("Enter the base value: "))
power = int(input("Enter the power value: "))
print("{} to the power of {} = {}".format(base,power,pow_rec(base,power)))
```
